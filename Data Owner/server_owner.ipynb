```python
# Import Statements
from PIL import Image, ImageOps
import numpy as np
import syft as sy
import pathlib
import random
import string
```

```python
# Hospital enters their data here
data_dir = pathlib.Path(input('Enter Data Folder Name: ')).with_suffix('')
image_size = (256, 256)
all_images = list(data_dir.glob('*/*.jpg'))
```

```python
image_count = len(list(data_dir.glob('*/*.jpg')))
print(image_count)
```

```python
# Folder Categories
categories = list(data_dir.glob('*'))
```

```python
# Get All Images
dataset_dict = {}
for (category_index, category) in enumerate(categories):
	category_dir = list(category.glob('*.jpg'))
	dataset_dict[category.name] = []
	for (image_index, image) in enumerate(category_dir):
		if image_index >= 1000: continue
		with Image.open(image) as img:
			resized_img = ImageOps.fit(img, image_size)
			dataset_dict[category.name].append(np.array(resized_img))
			print(f"Working on {image.name} in {category.name} (Image {image_index+1}/{len(category_dir)} of Category {category_index+1}/{len(categories)})")
	print("Converting Array to Numpy Array...")
	dataset_dict[category.name] = np.array(dataset_dict[category.name])
	print("Finished!")
```

```python
# Setting up Dataset
dataset = sy.Dataset(name='Skin Diseases')
dataset.set_description('Skin Disease Dataset')
dataset.add_contributor(
	role=sy.roles.UPLOADER,
	name=input('Enter Your Name: '),
	email=input('Enter Your Email: '),
	note=input('Enter Your Description: ')
)

dataset
```

```python
# Servers need the categories
skin_disease = sy.DataSubject(name='Skin Disease')
for disease in dataset_dict.keys():
	subject = sy.DataSubject(name=disease)
	skin_disease.add_member(subject)
result = client.data_subject_registry.add_data_subject(skin_disease)
data_subjects = client.data_subject_registry.get_all()
print(data_subjects)
```

```python
# Populating the project with the data
mock_permission = input('Allow Data Scientists To View Data (y/n): ').lower().strip() == 'y'
for (index, (category, nparray)) in enumerate(dataset_dict.items()):
	print(f"Creating Asset From {category}{' With Data Scientist Viewing Permission' if mock_permission else ''}")
	asset = sy.Asset(name=category)
	asset.set_description(f"This is all images collected on the following disease: {category}")
	asset.set_obj(nparray)
	asset.set_shape(nparray.shape)
	asset.set_mock(nparray, mock_is_real=True) if mock_permission else asset.no_mock()
	asset.add_data_subject(data_subjects[index+1])
	dataset.add_asset(asset)
```

```python
# Uploading Dataset to hospital servers
client.upload_dataset(dataset)
```

```python
def remedy_prompt():
	confirmation = input('Will You Allow Remedy To Use Your Dataset (y/n): ').lower().strip() == 'y'
	
	request = client.projects[0].requests[0]
	code = request.code
	
	all_pointers = {}
	for asset in request.code.assets:
		all_pointers[asset.name] = asset.data
	(print('Beginning Training...'), request.accept_by_depositing_result(code.unsafe_function(**all_pointers), force=True)) if confirmation else request.deny(reason=input('Please Tell Us Why You Denied: '))
```
